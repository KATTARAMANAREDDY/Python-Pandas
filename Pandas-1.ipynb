# Using slicing method to extract data

```python
# only isolating one column
# finding top 5 players names  

df['Player'].head()
```

```python
# isolating more than one column
# finding top 7 players avg

df[['Player','Ave']].head(7)
```

```python
# only extracting one row
# finding R Dravid total information

df[df.index==5]
```

```python
# extract more than one row 
# finding top 6 to 15 players information

df[df.index.isin(range(6,16))]
```

```python
# using loc[] and iloc[] (location and integer location) both are for specific for rows and both methods will give same information
# specific row
df.loc[3]
```

```python
# specific row
df.iloc[3]
```

```python
# start,stop

df.loc[0:5]
```

```python
# start,stop,step

df.iloc[0:7:2]
```
